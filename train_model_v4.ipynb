```python

from ultralytics import YOLO
import os

if __name__ == '__main__':
    # 사전 학습된 YOLOv8n 모델을 로드합니다.
    model = YOLO('yolov8n.pt')

    # 데이터셋 설정 파일 경로
    dataset_yaml_path = r'C:\Empty_Parking\dataset_v4\data_v4.yaml'

    if not os.path.exists(dataset_yaml_path):
        print(f"오류: 데이터셋 설정 파일을 찾을 수 없습니다. 경로: {dataset_yaml_path}")
        print("먼저 prepare_dataset_v4.py 스크립트를 실행하여 데이터셋을 준비해주세요.")
    else:
        print(f"데이터셋 설정 파일({dataset_yaml_path})을 사용하여 학습을 시작합니다.")
       
        results = model.train(
            data=dataset_yaml_path,
            epochs=150,
            imgsz=640,
            device=0
        )
        print("모델 학습이 완료되었습니다.")
```
