```python
import os
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')

input_dir = r"C:\Empty_Parking\bounding_box\P_test"

output_dir = r"C:\Empty_Parking\bounding_box\P_test_results"
os.makedirs(output_dir, exist_ok=True)

class_names = model.names
```

```python
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg"):
        # Full path to the image
        image_path = os.path.join(input_dir, filename)

        results = model(image_path)

        img = cv2.imread(image_path)

        for r in results:
            boxes = r.boxes
            for box in boxes:
                cls_id = int(box.cls[0])
                if class_names[cls_id] == 'car':
                    # Get the bounding box coordinates
                    x1, y1, x2, y2 = map(int, box.xyxy[0])
                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                    cv2.putText(img, "Occupied", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, img)

print("Processing complete.")
```
